```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import gmaps
from datetime import datetime

# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy

# Import the API key.
import requests
from config import weather_api_key
from config import g_key
```

```python
city_data_df = pd.read_csv("data/WeatherPy_vacation.csv")
city_data_df
```

```python
se_asia_df = city_data_df.loc[city_data_df['City'].isin(["Vung Tau", "Phan Thiet", "Nakhon Nayok", "Pak Chong"])]
se_asia_df
```

```python
se_asia_df_subset = se_asia_df[['City', 'Lat', 'Lng']]
se_asia_df_subset
```

```python
se_asia_tuple = [tuple(x) for x in se_asia_df_subset.to_numpy()]
se_asia_tuple
```

```python
locations = se_asia_df[["Lat", "Lng"]]
max_temp = se_asia_df["Max Temp"]
```

```python
import gmaps.datasets
gmaps.configure(api_key=g_key)

# Latitude-longitude pairs
Nakhon_Nayok = (14.2, 101.21)
Pak_Chong = (14.71, 101.42)
Vung_Tau = (10.35, 107.08)
Phan_Thiet = (10.93, 108.1)

fig = gmaps.figure()

info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Weather Description} and {Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
se_asia_markers = [info_box_template.format(**row) for index, row in se_asia_df.iterrows()]

marker_layer = gmaps.marker_layer(locations, info_box_content=se_asia_markers)

se_asia_travel = gmaps.directions_layer(
        Nakhon_Nayok, Phan_Thiet, waypoints=[Pak_Chong, Vung_Tau], show_markers=False,
        travel_mode = 'DRIVING')

fig.add_layer(se_asia_travel)
fig.add_layer(marker_layer)
fig
```
